```python
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open("synaps-delay-0.05.json", "r") as f:
    data = json.load(f)

with open("synaps-delay.json", "r") as f:
    data2 = json.load(f)

with open("synaps-delay-tet.json", "r") as f:
    data3 = json.load(f)
```

```python
df_1 = pd.DataFrame(data)
df_2 = pd.DataFrame(data2)
df_3 = pd.DataFrame(data3)
```

```python
df_1['train_loss']
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.plot(df_1['train_loss'], label='Train Loss (synaps-delay-0.05)')
ax1.plot(df_1['val_loss'], label='Validation Loss (synaps-delay-0.05)')
ax1.plot(df_2['train_loss'], label='Train Loss (synaps-delay-0.1)')
ax1.plot(df_2['val_loss'], label='Validation Loss (synaps-delay-0.1)')
# ax1.plot(df_3['train_loss'], label='Train Loss (synaps-delay-tet)')
# ax1.plot(df_3['val_loss'], label='Validation Loss (synaps-delay-tet)')
ax1.set_title('Training and Validation Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True)

ax2.plot(df_1['train_r2'], label='Train R2 (synaps-delay-0.05)')
ax2.plot(df_1['val_r2'], label='Validation R2 (synaps-delay-0.05)')
ax2.plot(df_2['train_r2'], label='Train R2 (synaps-delay-0.1)')
ax2.plot(df_2['val_r2'], label='Validation R2 (synaps-delay-0.1)')
# ax2.plot(df_3['train_r2'], label='Train R2 (synaps-delay-tet)')
# ax2.plot(df_3['val_r2'], label='Validation R2 (synaps-delay-tet)')
ax2.set_title('Training and Validation R2 Score')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('R2 Score')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
